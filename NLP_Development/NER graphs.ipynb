```python
import pandas as pd
import nltk
from bs4 import BeautifulSoup
import string
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer, word_tokenize
from nltk.tag import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
from nltk.tree import Tree
from nltk.chunk.api import ChunkParserI
import spacy
import en_core_web_sm
```

```python
tokens_df = pd.read_csv('tokens.csv')
tokens_df.head()
```

```python
#Apply NER for only one case - tokens
text = []
label = []
nlp = spacy.load('en_core_web_sm')
sentence = tokens_df['tokens'][0]
doc = nlp(sentence)
for ent in doc.ents:
    text.append(ent.text)
    label.append(ent.label_)
```

```python
#create a df
df = {'text':text, 'label':label}
df = pd.DataFrame(df)
#Count the label column
result = df.label.apply(lambda x: pd.value_counts(x.split(" "))).count()
#rename the column name
df1 = pd.DataFrame(result)
df1.rename( columns={0 : 'name'}, inplace=True )
#Create a new column 
df1['value'] = df1['name'].astype(int)
#Change the name column to be a str
df1['name'] = df1['name'].apply(str)
#remove the numbers of the name column
df1['name'] = df1['name'].str.replace(r'\d+','')
df1
```

```python
#create a barchart
ax = df1.plot.bar(title='MARBURY v. MADISON - 1803')
ax.set_xlabel("Label")
ax.set_ylabel("Total")
ax.figure.savefig('marVmad.png')
```
